# C4.5算法对鸢尾花数据集分类

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
```

```python
iris = pd.read_csv("iris.csv")
iris
```

```python
data = iris
sns.set_style("whitegrid")
antV = ['#1890FF', '#2FC25B', '#FACC14', '#223273', '#8543E0', '#13C2C2', '#3436c7', '#F04864']
# 绘制 Violinplot
f, axes = plt.subplots(2, 2, figsize=(8, 8), sharex=True)
sns.despine(left=True)
sns.violinplot(x='class', y='sepal_len', data=data, palette=antV, ax=axes[0, 0])
sns.violinplot(x='class', y='sepal_width', data=data, palette=antV, ax=axes[0, 1])
sns.violinplot(x='class', y='petal_len', data=data, palette=antV, ax=axes[1, 0])
sns.violinplot(x='class', y='petal_width', data=data, palette=antV, ax=axes[1, 1])
plt.savefig('analysis.png',format='png', dpi=300)
plt.show()
```

```python
categ_sepal_len = pd.cut(iris['sepal_len'],3)
catag_sepal_wid = pd.cut(iris['sepal_width'], 3)
catag_petal_len = pd.cut(iris['petal_len'], 3)
catag_petal_wid = pd.cut(iris['petal_width'], 3)
print(pd.value_counts(catag_petal_wid))
```

```python
import numpy as np
def DivideDataset(data):
    index = (np.random.permutation(len(data)))[0:15]
    data_test  = data.take(index)
    data_train = data.drop(index)
    datasets = [data_test, data_train]
    return datasets, data_test, data_train
```

```python
iris_data, iris_test, iris_train = DivideDataset(iris)
```
