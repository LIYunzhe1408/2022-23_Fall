### A. 数据导入
#### 1. 所有候选人信息
- 该文件为每个候选人提供一份记录，并显示候选人的信息、总收入、从授权委员会收到的转账、付款总额、给授权委员会的转账、库存现金总额、贷款和债务以及其他财务汇总信息。
- 数据字段描述详细:https://www.fec.gov/campaign-finance-data/all-candidates-file-description/

    ##### 关键字段说明

    * CAND_ID Candidate identification
    * CAND_NAME Candidate name
    * CAND_PTY_AFFILIATION Party affiliation
    * CAND_OFFICE_ST	Candidate state
    * CAND_OFFICE_DISTRICT	Candidate district
    * ......

`pandas.read_csv(file_path, sep, names)`
- file_path为文件路径；
- sep为分隔符，default为','；
- names为列名列表；

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 读取候选人信息，添加表头
candidates = pd.read_csv("weball22.txt", sep = '|',names=['CAND_ID','CAND_NAME','CAND_ICI','PTY_CD','CAND_PTY_AFFILIATION','TTL_RECEIPTS',
                                                          'TRANS_FROM_AUTH','TTL_DISB','TRANS_TO_AUTH','COH_BOP','COH_COP','CAND_CONTRIB',
                                                          'CAND_LOANS','OTHER_LOANS','CAND_LOAN_REPAY','OTHER_LOAN_REPAY','DEBTS_OWED_BY',
                                                          'TTL_INDIV_CONTRIB','CAND_OFFICE_ST','CAND_OFFICE_DISTRICT','SPEC_ELECTION','PRIM_ELECTION','RUN_ELECTION'
                                                          ,'GEN_ELECTION','GEN_ELECTION_PRECENT','OTHER_POL_CMTE_CONTRIB','POL_PTY_CONTRIB',
                                                          'CVG_END_DT','INDIV_REFUNDS','CMTE_REFUNDS'])
```

#### 2. 候选人委员会链接信息
- 该文件显示候选人的身份证号码、候选人的选举年份、联邦选举委员会选举年份、委员会识别号、委员会类型、委员会名称和链接标识号。
- 信息描述详细:https://www.fec.gov/campaign-finance-data/candidate-committee-linkage-file-description/

    ##### 关键字段说明

    * CAND_ID Candidate identification
    * CAND_ELECTION_YR Candidate election year
    * FEC_ELECTION_YR	FEC election year
    * CMTE_ID Committee identification
    * CMTE_TP	Committee type
    * CMTE_DSGN	Committee designation
        - *A = Authorized by a candidate*
        - *B = Lobbyist/Registrant PAC*
        - *D = Leadership PAC*
        - *J = Joint fundraiser*
        - *P = Principal campaign committee of a candidate*
        - *U = Unauthorized*
    * LINKAGE_ID	Linkage ID

```python
# 读取候选人和委员会的联系信息
ccl = pd.read_csv("ccl.txt", sep = '|',names=['CAND_ID','CAND_ELECTION_YR','FEC_ELECTION_YR','CMTE_ID','CMTE_TP','CMTE_DSGN','LINKAGE_ID'])
```

可能会出现一个候选人对应多个委员会的情况


需要分析的是候选人和捐款人的关联，因此需要将**没有候选人姓名**的`ccl.txt`表和候选人信息表关联

```python
# 关联两个表数据，将candidates表拼到ccl表上
ccl = pd.merge(ccl,candidates)

# 提取出所需要的列
ccl = pd.DataFrame(ccl, columns=[ 'CMTE_ID','CAND_ID', 'CAND_NAME','CAND_PTY_AFFILIATION'])
```

#### 3. 个人捐款档案信息 
- 本数据集只包含2022.6.29-2022.8.17的相关数据，如果需要更全数据可以通过数据来源中的地址下载。

- 该文件包含有关收到捐款的委员会、披露捐款的报告、提供捐款的个人、捐款日期、金额和有关捐款的其他信息。
- 信息描述详细:https://www.fec.gov/campaign-finance-data/contributions-individuals-file-description/

    ##### 关键字段说明

    * CMTE_ID 委员会ID
    * NAME 捐款人姓名
    * CITY 捐款人所在市
    * State 捐款人所在州
    * EMPLOYER 捐款人雇主/公司
    * OCCUPATION 捐款人职业
    * TRANSACTION_DT	Transaction date (MMDDYYYY)
    * TRANSACTION_AMT	Transaction amount

```python
# 读取个人捐赠数据，由于原始数据没有表头，需要添加表头
# 提示：读取本文件大概需要5-10s
individual_contribution = pd.read_csv('itcont_2022_20220629_20220817.txt', sep='|',names=['CMTE_ID','AMNDT_IND','RPT_TP','TRANSACTION_PGI',
                                                                                  'IMAGE_NUM','TRANSACTION_TP','ENTITY_TP','NAME','CITY',
                                                                                  'STATE','ZIP_CODE','EMPLOYER','OCCUPATION','TRANSACTION_DT',
                                                                                  'TRANSACTION_AMT','OTHER_ID','TRAN_ID','FILE_NUM','MEMO_CD',
                                                                                  'MEMO_TEXT','SUB_ID'])
```

```python
# 将候选人与委员会关系表ccl和个人捐赠数据表itcont合并，通过 CMTE_ID
# ccl表：候选人ID&姓名，委员会ID
# indi_con表：委员会ID，捐款信息
cand_cmte_indi =  pd.merge(ccl,individual_contribution)
# 提取需要的数据列
cand_cmte_indi = pd.DataFrame(cand_cmte_indi, columns=[ 'CAND_NAME','NAME', 'STATE','EMPLOYER','OCCUPATION',
                                           'TRANSACTION_AMT', 'TRANSACTION_DT','CAND_PTY_AFFILIATION'])
```

### B. 缺失值处理

`dataframe.info()`

```python
# 查看数据规模 多少行 多少列
data_scale = cand_cmte_indi.shape
# 查看整体数据信息，包括每个字段的名称、非空数量、字段的数据类型
cand_cmte_indi.info()
```

```python
#空值处理，统一填充 NOT PROVIDED
cand_cmte_indi['STATE'].fillna('NOT PROVIDED',inplace=True)
cand_cmte_indi['EMPLOYER'].fillna('NOT PROVIDED',inplace=True)
cand_cmte_indi['OCCUPATION'].fillna('NOT PROVIDED',inplace=True)

# 对日期TRANSACTION_DT列进行处理
cand_cmte_indi['TRANSACTION_DT'] = cand_cmte_indi['TRANSACTION_DT'] .astype(str)
# 将日期格式改为年月日  20220817
cand_cmte_indi['TRANSACTION_DT'] = [i[3:7]+i[0]+i[1:3] for i in cand_cmte_indi['TRANSACTION_DT'] ]

cand_cmte_indi.head(10)
```

## 信息熵公式计算喜好度概率

```python
import numpy as np
candidate_1 = 'SULLIVAN, DAN'
candidate_2 = 'DEMINGS, VAL'

#从所有数据中取出支持Sullivan的数据
info_candidate_1 = cand_cmte_indi[cand_cmte_indi['CAND_NAME']==candidate_1]
candidate_1_state = info_candidate_1.groupby('STATE').sum().sort_values("TRANSACTION_AMT", ascending=False)
q=np.array(candidate_1_state)
number_b=len(q)   #支持Sullivan的数量

# 从所有数据中取出支持Demings的数据
info_candidate_2 = cand_cmte_indi[cand_cmte_indi['CAND_NAME']==candidate_2]
candidate_2_state = info_candidate_2.groupby('STATE').sum().sort_values("TRANSACTION_AMT", ascending=False)
w=np.array(candidate_2_state) 
number_t=len(w)  #支持Demings州的数量


by_occupation=pd.pivot_table(cand_cmte_indi,index=['STATE'],values=['TRANSACTION_AMT'],aggfunc='sum')                                                                                                 
number=len(by_occupation)  #州的数量

# 统计各职业对Sullivan，Demings的捐款总数
candidate_1_state = info_candidate_1.groupby('OCCUPATION').sum().sort_values("TRANSACTION_AMT", ascending=False)
e=np.array(candidate_1_state)
candidate_2_state = info_candidate_2.groupby('OCCUPATION').sum().sort_values("TRANSACTION_AMT", ascending=False)  
r=np.array(candidate_2_state) 


by_occupation=pd.pivot_table(cand_cmte_indi,index=['OCCUPATION'],values=['TRANSACTION_AMT'],aggfunc='sum')                                                                                                 
number_j=len(by_occupation)


print('州的总数:{}'.format(number))
print('支持SULLIVAN, DAN州的数量:{}'.format(number_b))
print('支持DEMINGS, VAL州的数量:{}'.format(number_t))

print('职业的总数:{}'.format(number_j))
print('支持SULLIVAN, DAN的数量：{}'.format(len(e)))
print('支持DEMINGS, VAL的数量:{}'.format(len(r)))
```

## 热力图及每日曲线分析

```python
# 提取需要的数据列
c_itcont0 = pd.DataFrame(cand_cmte_indi, columns=[ 'CAND_NAME', 'TRANSACTION_AMT', 'TRANSACTION_DT'])
```

```python
# 已知所获得的捐款总额前两位的总统候选人分别是'BIDEN, JOSEPH R JR', 'TRUMP, DONALD J.'， 筛选出来
c_itcont1 = c_itcont0[c_itcont0['CAND_NAME'].isin(['SULLIVAN, DAN', 'HASSAN, MARGARET WOOD'])]
```

```python
# 因为同一天会有多笔捐款入帐，需要做分组求和
c_itcont2 = c_itcont1.groupby(['CAND_NAME', 'TRANSACTION_DT']).sum()
```

```python
# 因为我们要看总额的趋势图，所以要做累加
c_itcont3 = c_itcont2.groupby(['CAND_NAME', ]).cumsum()
c_itcont3
```

```python
# 整理下排列索引，这里用到了索引和列标题转换与转置操作
c_itcont4 = c_itcont3.unstack().T
c_itcont4
```

```python
c_itcont = cand_cmte_indi
first = c_itcont.groupby("CAND_NAME").sum().sort_values("TRANSACTION_AMT",ascending=False).index[1]
second = c_itcont.groupby("CAND_NAME").sum().sort_values("TRANSACTION_AMT",ascending=False).index[2]
x = list(c_itcont[c_itcont['CAND_NAME'] == first].groupby("TRANSACTION_DT")['TRANSACTION_AMT'].sum().index)
```

```python
y1 = list(c_itcont[c_itcont['CAND_NAME'] == first].groupby("TRANSACTION_DT")['TRANSACTION_AMT'].sum())
y2 = list(c_itcont[c_itcont['CAND_NAME'] == second].groupby("TRANSACTION_DT")['TRANSACTION_AMT'].sum())[:33]
y1
```

```python
fig = plt.figure(figsize=(15, 8), dpi=80)
plt.xticks(rotation=40)
plt.xlabel('date')
plt.ylabel('money')
plt.plot(x, y1, label=first, color='red')
plt.plot(x, y2, label=second, color='blue')
plt.legend(loc='upper left')
plt.grid(alpha=0.2)
plt.title('Overview of two major candidates')
```
